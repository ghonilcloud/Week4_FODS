```python
columns_to_convert = ['Postal Code', 'Model Year', 'Electric Range', 'Base MSRP', 'Legislative District', 'DOL Vehicle ID']

for col in columns_to_convert:
    df1_copy[col] = pd.to_numeric(df1_copy[col], errors='coerce')
    df2_copy[col] = pd.to_numeric(df2_copy[col], errors='coerce')

imputer = SimpleImputer(strategy='constant', fill_value=0)
df1_imputed = pd.DataFrame(imputer.fit_transform(df1_copy[columns_to_convert]), columns=columns_to_convert)
df2_imputed = pd.DataFrame(imputer.fit_transform(df2_copy[columns_to_convert]), columns=columns_to_convert)

df1_copy[columns_to_convert] = df1_imputed.astype(int)
df2_copy[columns_to_convert] = df2_imputed.astype(int)

columns_to_add = ['County', 'City', 'State', 'Make', 'Model', 'Electric Vehicle Type', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Vehicle Location', 'Electric Utility']

imputer = SimpleImputer(strategy='constant', fill_value="Unknown")
df1_imputed = pd.DataFrame(imputer.fit_transform(df1_copy[columns_to_add]), columns=columns_to_add)
df2_imputed = pd.DataFrame(imputer.fit_transform(df2_copy[columns_to_add]), columns=columns_to_add)

df1_copy[columns_to_add] = df1_imputed
df2_copy[columns_to_add] = df2_imputed
```
